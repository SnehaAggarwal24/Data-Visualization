```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Titanic dataset
titanic_data = sns.load_dataset('titanic')

# Drop rows with missing age or fare
line_data = titanic_data[['age', 'fare']].dropna().sort_values(by='age')

# Plot
plt.figure(figsize=(10, 5))
plt.plot(line_data['age'], line_data['fare'], color='blue')
plt.title('Line Graph of Fare by Age')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.grid(True)
plt.show()

```

```python
# Drop rows with missing age
age_data = titanic_data['age'].dropna()

# Plot
plt.figure(figsize=(8, 5))
plt.hist(age_data, bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram of Passenger Ages')
plt.xlabel('Age')
plt.ylabel('Number of Passengers')
plt.show()

```

```python
# Calculate class distribution
pclass_counts = titanic_data['pclass'].value_counts()
labels = ['First Class', 'Second Class', 'Third Class']
colors = ['gold', 'lightgreen', 'lightcoral']

# Plot
plt.figure(figsize=(6, 6))
plt.pie(pclass_counts, labels=labels, autopct='%1.1f%%', colors=colors)
plt.title('Passenger Class Distribution')
plt.show()
```

```python
# Drop rows with missing age
box_data = titanic_data[['pclass', 'age']].dropna()

# Plot
plt.figure(figsize=(8, 5))
sns.boxplot(x='pclass', y='age', data=box_data, palette='Set3')
plt.title('Boxplot of Age by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.show()
```
